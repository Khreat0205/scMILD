```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# GMM 파라미터 설정
mu1, mu2 = 0.2, 0.8
sigma1, sigma2 = 0.2, 0.2
pi1, pi2 = 0.7, 0.3

# 시각화를 위한 데이터 범위 설정
x = np.linspace(-1, 2, 1000)

# 각 가우시안 컴포넌트의 PDF 계산
pdf1 = pi1 * norm.pdf(x, mu1, sigma1)
pdf2 = pi2 * norm.pdf(x, mu2, sigma2)

# GMM의 PDF 계산
pdf_gmm = pdf1 + pdf2

# 결과 시각화
plt.figure(figsize=(8, 6))
plt.plot(x, pdf1, label=f'Gaussian 1 (mu={mu1}, sigma={sigma1})', linestyle='--')
plt.plot(x, pdf2, label=f'Gaussian 2 (mu={mu2}, sigma={sigma2})', linestyle='--')
plt.plot(x, pdf_gmm, label='Gaussian Mixture', linewidth=2)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Gaussian Mixture Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import torch
from sklearn.mixture import GaussianMixture

# 토이 데이터 생성
instance_label_pred_normed_np = np.array([[0.8], [0.9], [1.0], [0.1], [0.2], [0.3]])

# GMM 모델 학습
gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(instance_label_pred_normed_np)
print("GMM component means:", gmm.means_)

# 레이블 예측 및 component_order 설정
probabilities = gmm.predict_proba(instance_label_pred_normed_np)
component_labels = np.argmax(probabilities, axis=1)
if np.abs(gmm.means_[0] - gmm.means_[1]) < 0.3:
    component_labels[:] = 0
    gmm_component_order = [0, 0]
else:
    if gmm.means_[0] > gmm.means_[1]:
        gmm_component_order = [1, 0]
    else:
        gmm_component_order = [0, 1]
    component_labels = [gmm_component_order[label] for label in component_labels]

unique_labels, component_counts = np.unique(component_labels, return_counts=True)
print("Unique component labels:", unique_labels)
print("Component counts:", component_counts)

# pseudo_instance_label 처리
pseudo_instance_label_np = instance_label_pred_normed_np
component_labels = gmm.predict(pseudo_instance_label_np)
component_labels = [gmm_component_order[label] for label in component_labels]
pseudo_instance_label = torch.tensor(component_labels, dtype=torch.float)
print("Pseudo instance label:", pseudo_instance_label)

import numpy as np
from sklearn.mixture import GaussianMixture

# 토이 데이터 생성
instance_label_pred_normed_np = np.array([[0.1], [0.2], [0.3], [0.8], [0.9], [1.0]])

# GMM 모델 학습
gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(instance_label_pred_normed_np)
print("GMM component means:", gmm.means_)

# 레이블 예측 및 component_order 설정
probabilities = gmm.predict_proba(instance_label_pred_normed_np)
component_labels = np.argmax(probabilities, axis=1)
if np.abs(gmm.means_[0] - gmm.means_[1]) < 0.3:
    component_labels[:] = 0
    gmm_component_order = [0, 0]
else:
    if gmm.means_[0] > gmm.means_[1]:
        gmm_component_order = [1, 0]
    else:
        gmm_component_order = [0, 1]
    component_labels = [gmm_component_order[label] for label in component_labels]

unique_labels, component_counts = np.unique(component_labels, return_counts=True)
print("Unique component labels:", unique_labels)
print("Component counts:", component_counts)

# pseudo_instance_label 처리
pseudo_instance_label_np = instance_label_pred_normed_np
component_labels = gmm.predict(pseudo_instance_label_np)
component_labels = [gmm_component_order[label] for label in component_labels]
pseudo_instance_label = torch.tensor(component_labels, dtype=torch.float)
print("Pseudo instance label:", pseudo_instance_label)
```

```python
import numpy as np
from sklearn.cluster import KMeans
import torch

# 1차원 벡터 데이터를 두 개의 클러스터로 나누는 함수
def kmeans_clustering_1d(data, n_clusters=2):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(data)
    labels = kmeans.labels_
    return labels, kmeans.cluster_centers_

# 예제 데이터
data = np.array([[0.1], [0.2], [0.3], [0.8], [0.9], [1.0]])

# K-평균 클러스터링 수행
labels, cluster_centers = kmeans_clustering_1d(data)

# 결과 출력
print("Cluster labels:", labels)
print("Cluster centers:", cluster_centers)

# Tensor 변환
pseudo_instance_label = torch.tensor(labels, dtype=torch.float)
print("Pseudo instance label:", pseudo_instance_label)
```
